### <u>Histogram</u>
- Represented as bins
- Bins represent the data points in the range
- plt.hist(ele,bins=count)

```python
import numpy as np
std1=np.random.randint(18,45,100)
#std1
std2=np.random.randint(15,40,100)
plt.hist(std1,bins=7)
plt.xlabel('Age Category')
plt.ylabel('No of students')
plt.show()
```

```python
plt.hist(std2,bins=10,color='c')
```

```python
std1
```

```python
std2
```

```python
plt.hist(std2,bins=10,color='c',orientation='horizontal')
```

```python
vals=[num for num in range(10,50,5)]
plt.hist(std1,vals,histtype='bar',orientation='vertical',color='black')
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals,rwidth=0.7,histtype='bar',orientation='vertical',color=['g','r'])
plt.title("Multiple Histograms")
```

 **Pie Plot**

```python
#marking the regions
x=[15,25,30,24,6]
plt.pie(x)
plt.show()
```

```python
#adding labels
#Assigning labels of f to x
fb=['Average','Good','Outstanding','Excellent','Moderate']
plt.pie(x,labels=fb)
plt.show()
```

```python
# Adding colors to the pie chart
plt.pie(x,labels=fb,colors=['b','cyan','m','orange','pink'])
plt.title("Sort by Responses")
plt.show()
```

```python
# Assigning % to regions
plt.pie(x,labels=fb,colors=['b','cyan','m','orange','pink'],autopct="%.f%%")
plt.title("Sort by Responses")
plt.show()
```

```python
ou=[0,0,1,0,0]
plt.pie(x,labels=fb,colors=['m','cyan','r','orange','blue'],autopct="%.f%%",shadow=True,startangle=180,explode=ou)
plt.title("Sort by Responses")
plt.show()
```
